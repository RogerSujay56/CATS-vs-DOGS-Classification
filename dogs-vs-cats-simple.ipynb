```python
filenames=os.listdir('/kaggle/input/dogs-vs-cats/train/train/')
len(filenames)
```

```python
filenames[0][:3]
```

```python
classes=[]
for filename in filenames:
    classes.append(filename[:3]) #classes.append(filename.split('.')[0])
```

```python
classes[:5]
```

```python
df=pd.DataFrame({'NAME':filenames,'CLASS':classes})
```

```python
df.shape
```

```python
df.head()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
data_gen=ImageDataGenerator(shear_range=0.2,zoom_range=0.2,horizontal_flip=True,rescale=1/255)
```

```python
train_data = data_gen.flow_from_dataframe(df,'/kaggle/input/dogs-vs-cats/train/train/',
                                          x_col='NAME',y_col='CLASS',target_size=(224,224))
```

```python
train_data.class_indices
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Flatten,Dense
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(224,224,3)))
model.add(MaxPool2D())
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPool2D())
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPool2D())
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPool2D())
model.add(Flatten())
model.add(Dense(2,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',metrics=['Accuracy'])
```

```python
model.summary()
```

```python
model.fit_generator(train_data,epochs=5)
```

```python
model.save('mymodel.h5')
```
